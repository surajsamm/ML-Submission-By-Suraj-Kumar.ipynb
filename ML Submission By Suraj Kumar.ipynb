````python
# Chart - 2 visualization code
# Calculate number of bookings for each hotel type
# Calculate percentages
city_hotel_bookings = df[df['hotel'] == 'City Hotel'].shape[0]
resort_hotel_bookings = df[df['hotel'] == 'Resort Hotel'].shape[0]

# Calculate total bookings first
total_bookings = city_hotel_bookings + resort_hotel_bookings

city_hotel_percentage = (city_hotel_bookings / total_bookings) * 100
resort_hotel_percentage = (resort_hotel_bookings / total_bookings) * 100

print(f"Percentage of City Hotel bookings: {city_hotel_percentage:.2f}%")
print(f"Percentage of Resort Hotel bookings: {resort_hotel_percentage:.2f}%")
````

````python
#plot piechart from how many % or city hotel and resort hotel book plot piee chart
CityHotel = 66.70
ResortHotel = 33.30
labels = ['City Hotel', 'Resort Hotel']
sizes = [CityHotel, ResortHotel] # Add sizes array
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # explode 1st slice
fig1, ax1 = plt.subplots()
wedges, texts, autotexts = ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90, wedgeprops={'edgecolor': 'black'}) # Remove extra indent
ax1.axis('equal')
plt.title('Bookings Revenue: City Hotel vs Resort Hotel')

plt.legend(wedges, labels,
          title="Hotel Type",
          loc="upper left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
````

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

**Hypothesis 1:** There is a significant difference in average revenue between City Hotels and Resort Hotels.

**Hypothesis 2:** The proportion of canceled bookings differs significantly between City Hotels and Resort Hotels.

**Hypothesis 3:** Lead times for bookings differ significantly based on the customer type (Transient, Contract, Group, or Transient-Party).

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

**Hypothesis 1:** Average Revenue Difference between City Hotels and Resort Hotels
Null Hypothesis (H0): There is no significant difference in average revenue between City Hotels and Resort Hotels.



#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats

# Group data by hotel type
city_hotel_revenue = df[df['hotel'] == 'City Hotel']['revenue']
resort_hotel_revenue = df[df['hotel'] == 'Resort Hotel']['revenue']

# Perform independent samples t-test
t_stat, p_value = stats.ttest_ind(city_hotel_revenue, resort_hotel_revenue, nan_policy='omit')

print(f"T-statistic: {t_stat}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in average revenue between City Hotels and Resort Hotels.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average revenue between City Hotels and Resort Hotels.")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

In Hypothesis 1, the statistical test used to obtain the p-value is the independent samples t-test. This test compares the means of two independent groups (in this case, City Hotels and Resort Hotels) to determine if there is a statistically significant difference between them. The test calculates the t-statistic and the p-value, which indicates the probability of observing the data if the null hypothesis (no difference in means) is true.

##### Why did you choose the specific statistical test?

Answer Here.


The independent samples t-test was chosen for Hypothesis 1 because it is ideal for comparing the means of two independent groups (City Hotels and Resort Hotels) with continuous data (revenue). It assesses if there's a significant difference between the groups and is robust and interpretable for this analysis.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

**Hypothesis 2:** Proportion of Canceled Bookings between City Hotels and Resort Hotels **bold text**

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Create a contingency table
contingency_table = pd.crosstab(df['hotel'], df['is_canceled'])

# Perform chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2_stat}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: The proportion of canceled bookings is different for City Hotels and Resort Hotels.")
else:
    print("Fail to reject the null hypothesis: The proportion of canceled bookings is the same for City Hotels and Resort Hotels.")
````

**Hypothesis 3: Lead Time Difference Based on Customer Type**


#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Group data by customer type
lead_time_transient = df[df['customer_type'] == 'Transient']['lead_time']
lead_time_contract = df[df['customer_type'] == 'Contract']['lead_time']
lead_time_group = df[df['customer_type'] == 'Group']['lead_time']
lead_time_transient_party = df[df['customer_type'] == 'Transient-Party']['lead_time']

# Perform ANOVA
f_stat, p_value = stats.f_oneway(lead_time_transient, lead_time_contract, lead_time_group, lead_time_transient_party)

print(f"F-statistic: {f_stat}, P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in lead times between different customer types.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in lead times between different customer types.")
````

````python
# Remove URLs & Remove words and digits contain digits
def clean_text(text):
    # Remove URLs
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    text = url_pattern.sub('', text)

    # Remove words and digits containing digits
    text = re.sub(r'\b\w*\d\w*\b', '', text)  # Remove words with digits
    text = re.sub(r'\b\d+\b', '', text)       # Remove standalone digits

    return text

# Example usage:
text_with_urls_digits = "Check out my website at https://www.example.com. There are 123 apples and 456 oranges in the basket."
cleaned_text = clean_text(text_with_urls_digits)
print(cleaned_text)

````

````python
# Remove White spaces
def remove_whitespace(text):
    # Remove leading and trailing white spaces
    text = text.strip()

    # Replace multiple spaces, tabs, and newlines with a single space
    text = ' '.join(text.split())

    return text

# Example usage:
text_with_spaces = "   This  is   a    text   with \t  spaces\nand \nnewlines.   "
cleaned_text = remove_whitespace(text_with_spaces)
print(cleaned_text)
````

#### 6. Rephrase Text

````python
# Rephrase Text
def get_synonyms(word):
    synonyms = []
    for synset in wordnet.synsets(word):
        for lemma in synset.lemmas():
            synonyms.append(lemma.name())
    return synonyms

def rephrase_text(text):
    words = text.split()
    rephrased_words = []

    for word in words:
        # Get synonyms for the current word
        synonyms = get_synonyms(word)

        # Choose a random synonym if available
        if synonyms:
            rephrased_words.append(random.choice(synonyms))
        else:
            rephrased_words.append(word)

    # Join rephrased words into a single string
    rephrased_text = ' '.join(rephrased_words)

    return rephrased_text

# Example usage:
original_text = "Climate change is a pressing global issue that requires immediate action from governments and individuals alike."
rephrased_text = rephrase_text(original_text)
print("Original Text:")
print(original_text)
print("Rephrased Text:")
print(rephrased_text)
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
def stem_text(text):
    stemmer = PorterStemmer()
    tokens = nltk.word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    stemmed_text = ' '.join(stemmed_tokens)
    return stemmed_text

# Example usage:
text = "Stemming reduces words to their base forms, but it may not always generate meaningful words."
stemmed_text = stem_text(text)
print("Original Text:")
print(text)
print("\nStemmed Text:")
print(stemmed_text)
````
